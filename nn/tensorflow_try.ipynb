```python
import tensorflow as tf
import numpy as np
print(tf.__version__)
```

```python
from tensorflow.keras.layers import Dense
```

```python
x = np.array([[200.0, 70.0]])
layer_1 = Dense(units = 3, activation='sigmoid')
a1 = layer_1(x)
a1
```

```python
layer_2 = Dense(units = 1, activation='sigmoid')
a2 = layer_2(a1)
a2
```

```python
if a2 >= 0.5:
   print("True")
else:
   print('False')
```
