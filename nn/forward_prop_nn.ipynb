```python
def sigmoid(z):
   """
   Calculate the sigmoid function of a z score. Mostly used for logistic regression
   """
   return 1 / (1 + np.exp(-z))
```

```python
w1_1 = np.array([1,2])
b1_1 = np.array([-1])
z1_1 = np.dot(w1_1,x) + b1_1
a1_1 = sigmoid(z1_1)
a1_1
```

```python
w1_3 = np.array([5,-6])
b1_3 = np.array([2])
z1_3 = np.dot(w1_3,x) + b1_3
a1_3 = sigmoid(z1_3)
a1_3
```

```python
def Dense(a, W, b):
   """
   Dense is an implementation of a dense layer of a Neural Network(NN) construction
   """
   units = W.shape[1]
   a_out = np.zeros(units)
   for j in range(units):
      w = W[:, j]
      z = np.dot(w, a) + b[j]
      a_out[j] = sigmoid(z)
   return a_out
```
